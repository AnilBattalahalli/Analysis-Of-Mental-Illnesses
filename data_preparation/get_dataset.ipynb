```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_type = pd.read_excel("./mental_health_by_country.xlsx", "prevalence-by-mental-and-substa")
```

```python
df_type.dropna(inplace=True)
```

```python
df_type = df_type[df_type.Year <= 2017]
df_type = df_type[df_type.Year >= 1990]
```

```python
df_type.shape
```

```python
df_suicide = pd.read_excel("./mental_health_by_country.xlsx", "suicide-rates-vs-prevalence-of-")
```

```python
valid_years = [str(i) for i in range(1990, 2018)]
```

```python
df_suicide.dropna(inplace=True)
```

```python
df_suicide['Year'] = df_suicide["Year"].astype(int)
```

```python
df_gdp = pd.read_csv("GDP.csv")
```

```python
df_gdp = df_gdp.melt(df_gdp.columns[:2], 
        var_name="Year", 
        value_name="GDP")
```

```python
df_gdp = df_gdp[df_gdp.Year.str.contains('|'.join(valid_years))]
```

```python
df_gdp.sort_values(by=['Country ', 'Year'], inplace=True)
df_gdp.rename(columns={'Country ':'Country'}, inplace=True)
```

```python
a = df_gdp.values.tolist()
u_gdp = ["-".join([i[1],str(i[2])]) for i in a]
```

```python
a = df_type.values.tolist()
u_type = ["-".join([i[1],str(i[2])]) for i in a]
```

```python
a = df_suicide.values.tolist()
u_suicide = ["-".join([i[1],str(i[2])]) for i in a]
```

```python
df_gdp["cc"] = u_gdp
df_type["cc"] = u_type
df_suicide["cc"] = u_suicide
```

```python
df_gdp.set_index('cc', inplace=True)
df_type.set_index('cc', inplace=True)
df_suicide.set_index('cc', inplace=True)
```

```python
df_suicide = df_suicide[df_suicide.columns[3:]]
```

```python
df_gdp = df_gdp[df_gdp.columns[2:]]
```

```python
df_gdp = df_gdp[df_gdp.columns[1:]]
```

```python
df_grand = df_type.join(df_gdp, how='outer').join(df_suicide, how='outer')
```

```python
df_grand.to_csv("data_grand.csv")
```

```python
df_grand.shape
```

```python
df_grand.dropna().to_csv("grand_nona.csv")
```

```python
df_type.join(df_suicide, how='outer').to_csv("data_nogdp.csv")
```
